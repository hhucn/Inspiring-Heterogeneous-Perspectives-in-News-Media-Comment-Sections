```python
from sentence_transformers import SentenceTransformer
import pandas as pd
import torch
```

```python
model = SentenceTransformer('paraphrase-MiniLM-L12-v2')
```

```python
def read_article_from_dataset(path: str, filename: str) -> pd.DataFrame:
    return pd.read_csv(path + filename)[["articleID", "headline", "keywords"]]
```

```python
articles = read_article_from_dataset(path="", filename="ArticlesApril2017_DEV_LARGE.csv")
len(articles)
```

```python
print(articles.iloc[0])
```

```python
from tqdm import tqdm

embeddings = []

for article in tqdm(articles.itertuples()):
      embedding = model.encode(article.keywords, convert_to_tensor=True)
      embeddings.append({"ID": article.articleID,
                         "headline": article.headline,
                         "keywords": article.keywords,
                         "embedding": embedding})
```

```python
import torch
torch.save(embeddings, "embedded_articles_paraphrase-MiniLM-L12-v2.pt")
```
